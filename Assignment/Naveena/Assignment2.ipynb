#**3**. **Perform the below visualization**

**Univariate analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

data= pd.read_csv('/content/sample_data/excel.csv')

Numbercount = data['RowNumber']
 
Id = data['CustomerId']

plt.plot(Numbercount, Id, label='CustomerId')

plt.legend()

plt.savefig("plot.png")
plt.show()
```

**Bi-Variate analysis**

```python
import matplotlib.pyplot as plt 

Numbercount = data['RowNumber']
 
Id = data['CustomerId']
name = data['Surname']

plt.plot(Numbercount, Id, label='CustomerId')
plt.plot(Numbercount, name, label='Surname')

plt.legend()

plt.savefig("plot.png")
plt.show()
```

 **Multi-Variate Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/sample_data/excel.csv')

Numbercount = data['RowNumber']

Id = data['CustomerId']
name = data['Surname']
Score = data['CreditScore']
Geo = data['Geography']
Gd = data['Gender']
Age = data['Age']
Tu = data['Tenure']
Bal = data['Balance']
Nop = data['NumOfProducts']
Hc = data['HasCrCard']
ISA = data['IsActiveMember']
ES = data['EstimatedSalary']
Ex = data['Exited']

plt.plot(Numbercount, Id, label='CustomerId')
plt.plot(Numbercount, name, label='Surname')
plt.plot(Numbercount, Score, label='CreditScore')
plt.plot(Numbercount, Geo, label='Geography')
plt.plot(Numbercount, Gd, label='Gender')
plt.plot(Numbercount, Age, label='Age')
plt.plot(Numbercount, Tu, label='Tenure')
plt.plot(Numbercount, Bal, label='Balance')
plt.plot(Numbercount, Nop, label='NumOfProduct')
plt.plot(Numbercount, Hc, label='HasCrCard')
plt.plot(Numbercount, ISA, label='IsActiveMember')
plt.plot(Numbercount, ES, label='EstimatorySalary')
plt.plot(Numbercount, Ex, label='Exited')

plt.legend()

plt.savefig("plot.png")
plt.show()

```
